```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import time
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 加载数据
#ad_static_feature_df = pd.read_csv('ad_static_feature.csv')
#ad_operation_df = pd.read_csv('ad_operation.csv')
exposure_log_df = pd.read_csv('totalExpLog.csv')
exposure_log_df.info()
```

```python
Group_Exposure_Data = exposure_log_df.groupby(['year', 'month', 'day', 'ad_id', 'Ad_bid']).size().reset_index()
Group_Exposure_Data = Group_Exposure_Data.rename(columns={0: 'num_click'})
Group_Exposure_Data.info()
```

```python
Group_Exposure_Data.head()
```

```python
# 将曝光数据按照年月日 广告id和广告竞价删除重复的元素之后进行合并
Total_Exposure_Log_Data_one = exposure_log_df.drop_duplicates(subset=['year', 'month', 'day', 'ad_id', 'Ad_bid'] ,keep="first").reset_index(drop=True)
Clicks_of_Exposure_Data = pd.merge(Total_Exposure_Log_Data_one, Group_Exposure_Data,on=['year', 'month', 'day', 'ad_id', 'Ad_bid'])
Clicks_of_Exposure_Data = Clicks_of_Exposure_Data.reset_index(drop = True)
Clicks_of_Exposure_Data.info()
```

```python
Clicks_of_Exposure_Data.head()
```

```python
# 删除测试集中没有的相关属性 并将结果进行保存
Clicks_of_Exposure_Data.drop('Ad_Request_id', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('Ad_Request_Time', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('user_id', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('Ad_material_size', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('Ad_pctr', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('Ad_quality_ecpm', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('Ad_total_Ecpm', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('year', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('month', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('day', axis=1, inplace=True)
Clicks_of_Exposure_Data.drop('Ad_pos_id', axis=1, inplace=True)
#Clicks_of_Exposure_Data.drop('Ad_bid', axis=1, inplace=True)
Clicks_of_Exposure_Data.info()
```

```python
Clicks_of_Exposure_Data.head(100)
```

```python

Clicks_of_Exposure_Data.to_csv('Clicks_of_Exposure_Data.csv', index=False)
 
# 将曝光日志按照ID和静态广告数据进行拼接操作
Ad_Static_Data = pd.read_csv('ad_static_feature.csv')
Ad_Static_Data.drop('Commodity_id', axis=1, inplace=True)
Ad_Static_Data.drop('Ad_account_id', axis=1, inplace=True)
Ad_Static_Data.drop('Creation_time', axis=1, inplace=True)

Merce_Ad_Static_and_Exposure_Data = pd.merge(Clicks_of_Exposure_Data, Ad_Static_Data, on=['ad_id'])
 
# 读取广告操作数据集并拼接数据集
Op_Ad_Data = pd.read_csv('ad_operation.csv').drop_duplicates(['ad_id'])
Op_Ad_Data.drop('Create_modify_time', axis=1, inplace=True)
 
Dataset_For_Train = pd.merge(Op_Ad_Data, Merce_Ad_Static_and_Exposure_Data, on=['ad_id'])
Dataset_For_Train = Dataset_For_Train.drop_duplicates().reset_index(drop=True)
Dataset_For_Train.to_csv('Dataset_For_Train.csv', index=False)
Dataset_For_Train.info()
```
